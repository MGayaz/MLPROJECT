```python
# plotting a bar chart for Gender and it's count
plt.figure(figsize=(6, 4))  # Optional: Set the figure size
df['Gender'].value_counts().plot(kind='bar',color=['blue','orange'])

plt.title('Count Plot of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# plotting a bar chart for gender vs total amount
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(6, 4))
plt.bar(sales_gen['Gender'],sales_gen['Amount'],color=['blue','orange'])
plt.title('Gen vs Total Amount')
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.show()
```

```python
plt.figure(figsize=(6, 4))  # Optional: Set the figure size
df['Age Group'].value_counts().plot(kind='bar',color=['blue','orange'])

plt.title('Count Plot of Gender')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(6, 4))
plt.bar(sales_age['Age Group'],sales_age['Amount'],color=['blue','orange'])
plt.title('Amount vs Age Group')
plt.xlabel('Amount')
plt.ylabel('Age Group')
plt.show()
```

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
plt.figure(figsize=(15, 4))
plt.bar(sales_state['State'],sales_state['Orders'],color=['blue','orange'])
plt.title('Orders from top 10 states')
plt.xlabel('State')
plt.ylabel('Orders')
plt.show()
```

```python
# total amount/sales from top 10 states
sales_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
plt.figure(figsize=(15,4))
plt.bar(sales_states['State'],sales_states['Amount'],color=['blue','orange'])
plt.title('sales from top 10 states')
plt.xlabel('State')
plt.ylabel('Orders')
plt.show()
```

```python
#From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
```

```python
#Marital Status
plt.figure(figsize=(15, 4))  # Optional: Set the figure size
df['Marital_Status'].value_counts().plot(kind='bar',color=['blue','orange'])

plt.title('Marital Status Plot')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
plt.figure(figsize=(15, 4))  # Optional: Set the figure size
df['Occupation'].value_counts().plot(kind='bar')

plt.title('Occupation Bar')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.show()
```

```python
sales_occup = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(15,4))
plt.bar(sales_occup['Occupation'],sales_occup['Amount'],color=['blue','orange'])
plt.title('Occupation Vs Amount')
plt.xlabel('Occupation')
plt.ylabel('Amoount')
plt.show()
```

```python
#From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

```python
plt.figure(figsize=(15, 4))  # Optional: Set the figure size
df['Product_Category'].value_counts().plot(kind='bar',color=['blue','orange'])

plt.title('Product Bar')
plt.xlabel('Product_Category')
plt.ylabel('Count')
plt.show()
```

```python
sales_prod = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(15,4))
plt.bar(sales_prod['Product_Category'],sales_prod['Amount'],color=['blue','orange'])
plt.title('Product Vs Amount')
plt.xlabel('Product Category')
plt.ylabel('Amoount')
plt.show()
```

```python
#From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
```

```python
sales_id = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
plt.figure(figsize=(15,4))
plt.bar(sales_id['Product_ID'],sales_id['Orders'],color=['blue','orange'])
plt.title('Product Vs Orders')
plt.xlabel('Product ID')
plt.ylabel('Orders')
plt.show()
```

```python
#Conclusion:
#Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
```
