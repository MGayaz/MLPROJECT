```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from  sklearn.metrics import r2_score
import os
```

```python
os.chdir("C:\\Users\\ACER\\Downloads\\Batch 75 Day12\\Batch 75 Day12")
df=pd.read_csv("Height_Weight_Dataset.csv")
print(df)
```

```python
x=df.iloc[:,0:1].values
print(x)
```

```python
y=df.iloc[:,1].values
print(y)
```

```python
plt.plot(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
print(x.shape)
print(y.shape)
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt
import matplotlib.pyplot as plt
```

```python
rmse_value=[]
for k in range (20):
    k=k+1
    model=neighbors.KNeighborsRegressor(n_neighbors=k)
    model.fit(x_train,y_train)
    pred=model.predict(x_train)
    errors=sqrt(mean_squared_error(pred,y_train))
    rmse_value.append(errors)
    print('RMSE value for k= ' , k , 'is:', errors)
```

```python
plt.plot(range(1,21),rmse_value)
plt.show()
```

```python
accu_value=[]
for k in range (20):
    k=k+1
    model=neighbors.KNeighborsRegressor(n_neighbors=k)
    model.fit(x_train,y_train)
    pred=model.predict(x_train)
    accuracy=r2_score(pred,y_train)
    accu_value.append(accuracy)
    print('Accuracy value for k= ' , k , 'is:', accuracy)
```

```python
plt.plot(range(1,21),accu_value)
plt.show()
```
