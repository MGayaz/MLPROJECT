```python
df=pd.get_dummies(df,columns=['3D_available','Genre'],drop_first=True)
display (df.head())
```

```python
x=df.loc[:, df.columns!='Collection']
display(x)
```

```python
y = df['Collection']
display(y)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_x.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
print(x_train.shape)
print(y_train.shape)
print(x.shape)
print(y.shape)
```

```python
from sklearn.svm import SVR
svr_poly = SVR(kernel ='poly', C= 1000)
svr_poly .fit (x_train,y_train)
print (svr_poly)
```

```python
print (y_test )
```

```python
svr_poly_pred=svr_poly.predict(x_train)
print(svr_poly_pred)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
accuracy=r2_score(svr_poly_pred,y_train)
rms=mean_squared_error(svr_poly_pred,y_train)
print(accuracy)
print(rms)
```

```python
from sklearn.svm import SVR
svr_sigm = SVR(kernel ='sigmoid', C= 1000)
svr_sigm .fit (x_train,y_train)
print (svr_sigm)
```

```python
svr_sigm_pred=svr_sigm.predict(x_train)
print(svr_sigm_pred)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
accuracy=r2_score(svr_sigm_pred,y_train)
rms=mean_squared_error(svr_sigm_pred,y_train)
print(accuracy)
print(rms)
```

```python
from sklearn.svm import SVR
svr_rbf = SVR(kernel ='rbf', C= 1000)
svr_rbf .fit (x_train,y_train)
print (svr_rbf)
```

```python
svr_rbf_pred=svr_rbf.predict(x_train)
print(svr_rbf_pred)
```
